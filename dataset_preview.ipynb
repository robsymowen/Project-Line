```python
import os
from torchvision.datasets import ImageNet

styles = ['anime_style', 'contour_style', 'opensketch_style']

def get_dataset(style_name, split, transform=None):
    assert style_name in styles, f"style_name must be one of {styles}, got {style_name}"
    assert split in ['train', 'val'], f"split must be 'train' or 'val', got {split}"
    root_directory = os.path.join(os.environ['SHARED_DATA_DIR'], 'imagenet1k-line', f"imagenet1k-{style_name}")
    dataset = ImageNet(root_directory, split=split, transform=transform)
    if split=="val":
        assert len(dataset)==50000, f"Expected 50000 val images, got {len(dataset)}"
    else:
        assert len(dataset)==1281167, f"Expected 1281167 val images, got {len(dataset)}"
    return dataset
```

# verify that the datasets align

## val datasets

```python
from pathlib import Path
```

```python
root_directory = os.path.join(os.environ['SHARED_DATA_DIR'], 'imagenet1k-256')
dataset1 = ImageNet(root_directory, "val")
dataset1
```

```python
dataset2 = get_dataset("anime_style", "val")
# dataset2 = get_dataset("contour_style", "val")
# dataset2 = get_dataset("opensketch_style", "val")
dataset2
```

```python
imgs1 = sorted([(os.path.join(Path(f).parent.name,Path(f).stem),label) for f,label in dataset1.imgs])
imgs1[0:10]
```

```python
imgs2 = sorted([(os.path.join(Path(f).parent.name,Path(f).stem.replace("_out","")),label) for f,label in dataset2.imgs])
imgs2[0:10]
```

```python
imgs1==imgs2
```
